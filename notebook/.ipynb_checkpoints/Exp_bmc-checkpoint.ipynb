```python
def plot_figs(res, worsts, X, yrange, xlabel, ylabel, name, is_scale, is_int=False):

    X_axis = np.arange(len(X))
    ax = plt.subplot(111)
    # Hide the right and top spines
    ax.spines['right'].set_visible(False)
    ax.spines['top'].set_visible(False)
    X_axis_list = [X_axis-bar_width * 3, X_axis-bar_width * 2, X_axis-bar_width , X_axis, X_axis + bar_width
             , X_axis+bar_width * 2, X_axis+bar_width * 3]
    
    for i in range(len(res)):
        b1 = ax.bar(X_axis_list[i], res[i], bar_width, label = labels[i], color=colors[i],alpha=alpha, hatch=hatchs[i])
        if is_int:
            ax.bar_label(b1, fmt='%d', padding=padding, fontsize=bar_label_fontsize, rotation=bar_label_rotation)
        else:
            ax.bar_label(b1, fmt='%.2f', padding=padding, fontsize=bar_label_fontsize, rotation=bar_label_rotation)
                
        

#     b1 = ax.bar(X_axis-bar_width * 2, hr, bar_width, label = labels[0], color=colors[0],alpha=alpha)
#     b2 = ax.bar(X_axis-bar_width, zr, bar_width, label = labels[1], color=colors[1],alpha=alpha)
#     b3 = ax.bar(X_axis, bcm_rl, bar_width, label = labels[2], color=colors[2],alpha=alpha)
#     b4 = ax.bar(X_axis+bar_width, quilts, bar_width, label = labels[3], color=colors[3],alpha=alpha)
#     b5 = ax.bar(X_axis+bar_width*2, rstar, bar_width, label = labels[4], color=colors[4],alpha=alpha)

#     ax.bar_label(b1, fmt='%.3f', padding=5, fontsize=bar_label_fontsize, rotation=bar_label_rotation)
#     ax.bar_label(b2, fmt='%.3f', padding=5, fontsize=bar_label_fontsize, rotation=bar_label_rotation)
#     ax.bar_label(b3, fmt='%.3f', padding=5, fontsize=bar_label_fontsize, rotation=bar_label_rotation)
#     ax.bar_label(b4, fmt='%.3f', padding=5, fontsize=bar_label_fontsize, rotation=bar_label_rotation)
#     ax.bar_label(b5, fmt='%.3f', padding=5, fontsize=bar_label_fontsize, rotation=bar_label_rotation)

    plt.xticks(X_axis, X)
    plt.ylim(yrange)

    if xlabel != "":
        plt.xlabel(xlabel, fontsize=26)
    plt.ylabel(ylabel, fontsize=26)
    if is_scale:
        plt.yscale('log')
    ax.xaxis.set_label_coords(.5, -0.1)
    plt.subplots_adjust(bottom=0.1)

    plt.legend(loc='upper right', ncol=4, labelspacing=0.1, columnspacing=0.5, borderpad=0, handlelength=1, fancybox=True, framealpha=0)

    plt.savefig('../../../../应用/Overleaf/'+floder+name,format='pdf',bbox_inches = 'tight')    
    # plt.savefig('../../应用/Overleaf/'+floder+'/figs/2_d_pattern.pdf',format='pdf',bbox_inches = 'tight')    
    plt.show()
```

```python
def get_ratio(bcm_rl, quilts, hr, zr, lo, rstar, rsmi, optimal):
    zr = [zr[i] / optimal[i] for i in range(len(zr))]
    hr = [hr[i] / optimal[i] for i in range(len(zr))]
    bcm_rl = [bcm_rl[i] / optimal[i] for i in range(len(zr))]
    quilts = [quilts[i] / optimal[i] for i in range(len(zr))]
    rstar = [rstar[i] / optimal[i] for i in range(len(zr))]
    lo = [lo[i] / optimal[i] for i in range(len(zr))]
    rsmi = [rsmi[i] / optimal[i] for i in range(len(zr))]
    res = [bcm_rl, quilts, hr, zr, lo, rstar, rsmi]
    return res
```

```python
def read_ratio(file_name):
    res = []
    with open(file_name) as f:
        lines = f.readlines()
        for line in lines:
            ratio = line.split(",")[-1].strip()
            if ratio == "":
                continue
            res.append(float(ratio))
    return res
```

```python
def get_quantile(arr):
    quantiles = [0, .25, .50, .75, 1]
    res = []
    for quantile in quantiles:
        res.append(np.quantile(arr, quantile))
    return res
```

```python
def plot_box_by_dataset(dataset):
    max_val = 1
    hc = read_ratio("./box_plot_ratios/" + dataset + "/HC.txt")
    hc.sort()
    max_val = max(max_val, max(hc))
    zc = read_ratio("./box_plot_ratios/" + dataset + "/ZC.txt")
    zc.sort()
    
    max_val = max(max_val, max(zc))
    bmc_lr = read_ratio("./box_plot_ratios/" + dataset + "/LBMC.txt")
    bmc_lr.sort()

    max_val = max(max_val, max(bmc_lr))
    
    quilts = read_ratio("./box_plot_ratios/" + dataset + "/QUILTS.txt")
    quilts.sort()
    '||',
    max_val = max(max_val, max(quilts))
    

    rstar = read_ratio("./box_plot_ratios/" + dataset + "/RStar.txt")
    rstar.sort()
    
    max_val = max(max_val, max(rstar))
    

    lc = read_ratio("./box_plot_ratios/" + dataset + "/LC.txt")
    lc.sort()
    max_val = max(max_val, max(lc))
    
    data=[]
    data.append(get_quantile(bmc_lr))
    data.append(get_quantile(quilts))
    data.append(get_quantile(hc))
    data.append(get_quantile(zc))
    data.append(get_quantile(lc))
    data.append(get_quantile(rstar))
    
    if dataset == "OSM":
        rsmi = read_ratio("./box_plot_ratios/" + dataset + "/RSMI.txt")
        rsmi.sort()
        max_val = max(max_val, max(rsmi))
        
        data.append(get_quantile(rsmi))
    plot_box(data, "Method", "Block access ratio", [1,max_val], True, "/figs/box_plot_"+dataset+".pdf", dataset)
```
